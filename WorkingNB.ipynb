```python
import pandas as pd
import uuid
```

```python
dfs = []
for i in range(2014,2024):
    dfs.append(pd.read_csv("src/acestor/tmpdir/" + str(i) + ".csv"))

df = pd.concat(dfs, ignore_index=True)
```

```python
df = df[df["ulb_code"].isnull()].reset_index(drop=True)
```

```python
df['metadata.ID'] = [uuid.uuid4() for _ in range(len(df.index))]
```

```python
df.columns
```

```python
todrop = ["type", "ulb_code","zone_name",'lat','lng','case_type','year','ward_number']
```

```python
columns = ["type", "metadata.primaryDate", "event.symptomOnset.date", 
           "event.test.sampleCollectionDate","event.test.resultDate", "location.admin1.ID", "location.admin2.ID",
           "location.admin3.ID", "ulb_code","location.admin4.ID","zone_name",
           "ward_number", "location.healthcenter.PHC", "location.healthcenter.subcenter", "lat","lng", "demographics.age", "demographics.gender",
           "event.test.type","case_type","location.admin2.name","location.admin3.name","location.admin4.name","year", "metadata.ID"
]
```

```python
df.columns = columns
df.drop(todrop, axis = "columns", inplace = True)
```

```python
dfprime = df.copy()
#df.rename({"metadata.date": "metadata.primaryDate"}, axis = "columns", inplace=True)
```

```python
df
```

```python
df["location.country.ID"] = "country_IN"
df["location.country.name"] = "INDIA"

df["location.admin.hierarchy"] = "revenue"

df["location.admin1.ID"] = df["location.admin1.ID"].fillna(-1).astype(int)
df["location.admin1.ID"] = "state_" + df["location.admin1.ID"].astype(str)
df["location.admin1.name"] = "KARNATAKA"

df["location.admin2.ID"] = df["location.admin2.ID"].fillna(-1).astype(int)
df["location.admin2.ID"] = "district_" + df["location.admin2.ID"].astype(str)

df["location.admin3.ID"] = df["location.admin3.ID"].fillna(-1).astype(int)
df["location.admin3.ID"] = "subdistrict_" + df["location.admin3.ID"].astype(str)

df["location.admin4.ID"] = df["location.admin4.ID"].fillna(-1).astype(int)
df["location.admin4.ID"] = "village_" + df["location.admin4.ID"].astype(str)
```

```python
df["metadata.primaryDate"] = pd.to_datetime(df["metadata.primaryDate"])

df["event.symptomOnset"] = True
df["event.symptomOnset.date"] = pd.to_datetime(df["event.symptomOnset.date"])

df["event.test"] = True
df["event.test.sampleCollectionDate"] = pd.to_datetime(df["event.test.sampleCollectionDate"])
df["event.test.type"] = df["event.test.type"].replace("IgM", "IgM ELISA").replace("NS1", "NS1 Ag ELISA").fillna("Unknown")
df["event.test.resultDate"] = pd.to_datetime(df["event.test.resultDate"])
df["event.test.result"] = True
```

```python
final_cols = ["metadata.ID","metadata.primaryDate",
              
              "demographics.age", "demographics.gender",

              "location.country.ID", "location.country.name",
                "location.admin.hierarchy",
              "location.admin1.ID", "location.admin1.name", 
              "location.admin2.ID", "location.admin2.name",
              "location.admin3.ID", "location.admin3.name", 
              "location.admin4.ID", "location.admin4.name",
              "location.healthcenter.PHC", "location.healthcenter.subcenter",

              "event.symptomOnset", "event.symptomOnset.date",

              "event.test", "event.test.sampleCollectionDate",
              "event.test.type", "event.test.resultDate","event.test.result"
              ]
```

```python
df = df[final_cols]
```

```python
df.to_csv("karnataka_linelist.csv", index=False)
```
